```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('penguins.csv')

print("企鵝數量:", len(df))
print("平均體重:", df['body_mass_g'].mean())
print("各物種數量:\n", df['species'].value_counts())
print("去掉 NaN:", df.dropna().shape)

df.plot(x='species', y='body_mass_g', kind='bar')
plt.savefig('penguins_analysis.png')
print("✅ Done")
```
